#Theory Questions and Answers

1.  What is hypothesis testing in statistics?
  - Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. It involves formulating two competing statements (hypotheses) and using sample evidence to decide whether to reject the null hypothesis. The process helps determine if observed differences are statistically significant or due to random chance.

2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?
  - Null Hypothesis (H₀): A statement that assumes no effect, no difference, or no relationship in the population. It represents the "status quo."

  - Alternative Hypothesis (H₁ or Ha): A statement that contradicts the null hypothesis, suggesting there is an effect, difference, or relationship.

  - Example:

    - H₀: The average height of students = 170 cm

    - H₁: The average height of students ≠ 170 cm

3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
  - The significance level (α) is the probability of rejecting the null hypothesis when it is actually true (Type I error). Common choices are 0.05 (5%) or 0.01 (1%).

  - If the p-value ≤ α, we reject H₀.

  - If the p-value > α, we fail to reject H₀.

  - It sets the threshold for how strong the evidence must be to reject the null hypothesis.

4. What are Type I and Type II errors? Give examples of each.
  - Type I Error (False Positive): Rejecting H₀ when it is actually true.
    - Example: A medical test indicates a patient has a disease when they are healthy.

  - Type II Error (False Negative): Failing to reject H₀ when H₁ is true.
    - Example: A medical test fails to detect a disease when the patient actually has it.

5. What is the difference between a Z-test and a T-test? Explain when to use
each.
  - Z-test: Used when population variance is known and sample size is large (n > 30).

  - T-test: Used when population variance is unknown and sample size is small (n ≤ 30).

  - Rule of thumb:

    - Use Z-test if σ (population standard deviation) is known.

    - Use T-test if σ is unknown and estimated using sample standard deviation.



```python
#6. Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

#(Include your Python code and output in the code box below.)
#Hint: Generate random number using random function.

import numpy as np
import matplotlib.pyplot as plt


n, p = 10, 0.5
data = np.random.binomial(n, p, 1000)


plt.hist(data, bins=range(0, n+2), edgecolor="black", density=True)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

```python
#7. Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
#sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        #50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        #50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                        #50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu0 = 50
xbar = np.mean(sample_data)
sigma = np.std(sample_data, ddof=1)
n = len(sample_data)


z = (xbar - mu0) / (sigma / np.sqrt(n))


p_value = 2 * (1 - norm.cdf(abs(z)))

print("Sample Mean:", xbar)
print("Z-statistic:", z)
print("p-value:", p_value)
```

```python
#8.  Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats


data = np.random.normal(loc=100, scale=15, size=500)


mean = np.mean(data)
sem = stats.sem(data)
ci = stats.norm.interval(0.95, loc=mean, scale=sem)


plt.hist(data, bins=30, edgecolor="black", density=True)
plt.axvline(ci[0], color="red", linestyle="dashed", label="95% CI Lower")
plt.axvline(ci[1], color="blue", linestyle="dashed", label="95% CI Upper")
plt.axvline(mean, color="green", label="Mean")
plt.title("Normal Distribution with 95% Confidence Interval")
plt.legend()
plt.show()

print("Mean:", mean)
print("95% Confidence Interval:", ci)
```

```python
#9.  Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent
# in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore


data = np.random.normal(50, 10, 200)


z_scores = zscore(data)


plt.hist(z_scores, bins=20, edgecolor="black", density=True)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()

print("First 10 Z-scores:", z_scores[:10])
```
